```python
import seaborn as sns
df=sns.load_dataset('tips')
df
```

```python
#corelation
df.corr()
```

```python
#heatmap
sns.heatmap(df.corr())
```

```python
#jointpolt
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.jointplot(x='tip',y='size',data=df,kind='reg')
```
